Now the output I actually want will be over $M$ runs of the SSA with all the values at steady state. To ensure steady state across all the moments I should take 10 times the maximal time scale associated with the dynamical process. 

```julia
tdeg = 1;
texpt(kA,kB,kC,S,R) = 1/kC;
tinit(kA,kB,kC,S,R) = (1/kB)+(S-1)/kA;
```

```julia
pars1 = [1,10,1]; hypars1 = [2,2];
```

```julia
t2 = texpt(pars1...,hypars1...)
```

```julia
t3 = tinit(pars1...,hypars1...)
```

```julia
M = 50
op = SSA(M, pars1, hypars1, maximum([1,t2,t3])*10, maximum([1,t2,t3])*10)
```

```julia
nuc_pop = op[1][:,2]
nuc_pop_m = mean(op[1][:,2])
```

```julia
meanN(pars1...,hypars1...)
```

```julia
cyto_pop = op[2][:,2]
cytp_pop_m = mean(op[2][:,2])
```

```julia
meanC(pars1...,hypars1...)
```
