# Histograms for all replicates

```julia
using Random, Plots, LaTeXStrings, Distributions, StatsBase, DataFrames, CSV, Plots.PlotMeasures, JLD2, GpABC
```

```julia
Plots.theme(:dao)
Plots.gr_cbar_width[] = 0.01;
```

```julia
pal = cgrad(:tol_muted);
cpal = [pal[i] for i in 1/10:1/10:1.0]
population_colors = cpal
```

```julia
samp1 = load("./jld2-res/S3R3_mad1WTGFP_samp1.jld2", "ABC")
samp2 = load("./jld2-res/S3R3_mad1WTGFP_samp2.jld2", "ABC")
samp3 = load("./jld2-res/S3R3_mad1WTGFP_samp3.jld2", "ABC")
samp4 = load("./jld2-res/S3R3_mad1WTGFP_samp4.jld2", "ABC");
```

```julia
pop1 = samp1.population[end]
pop2 = samp2.population[end]
pop3 = samp3.population[end]
pop4 = samp4.population[end];
```

```julia
kA1 = pop1[:,1]; kB1 = pop1[:,2]; kC1 = pop1[:,3];
kA2 = pop2[:,1]; kB2 = pop2[:,2]; kC2 = pop2[:,3];
kA3 = pop3[:,1]; kB3 = pop3[:,2]; kC3 = pop3[:,3];
kA4 = pop4[:,1]; kB4 = pop4[:,2]; kC4 = pop4[:,3];
```

```julia
kAm1 = median(kA1); kBm1 = median(kB1); kCm1 = median(kC1)
kAm2 = median(kA2); kBm2 = median(kB2); kCm2 = median(kC2)
kAm3 = median(kA3); kBm3 = median(kB3); kCm3 = median(kC3)
kAm4 = median(kA4); kBm4 = median(kB4); kCm4 = median(kC4);
```

```julia
kAlq1 = quantile(kA1,0.25); kBlq1 = quantile(kB1,0.25); kClq1 = quantile(kC1,0.25);
kAlq2 = quantile(kA2,0.25); kBlq2 = quantile(kB2,0.25); kClq2 = quantile(kC2,0.25);
kAlq3 = quantile(kA3,0.25); kBlq3 = quantile(kB3,0.25); kClq3 = quantile(kC3,0.25);
kAlq4 = quantile(kA4,0.25); kBlq4 = quantile(kB4,0.25); kClq4 = quantile(kC4,0.25);
```

```julia
kAuq1 = quantile(kA1,0.75); kBuq1 = quantile(kB1,0.75); kCuq1 = quantile(kC1,0.75);
kAuq2 = quantile(kA2,0.75); kBuq2 = quantile(kB2,0.75); kCuq2 = quantile(kC2,0.75);
kAuq3 = quantile(kA3,0.75); kBuq3 = quantile(kB3,0.75); kCuq3 = quantile(kC3,0.75);
kAuq4 = quantile(kA4,0.75); kBuq4 = quantile(kB4,0.75); kCuq4 = quantile(kC4,0.75);
```

## Replicate 1

```julia
a1 = stephist(kA1, fill = true, fillalpha = 0.5, color = cpal[1], label = :none, norm = true, titlelocation = :left)
plot!(grid = false, minorgrid = false, legend = :none, size = (450,400), ylims = [0,2.0], xlims = [2,6])
plot!([kAm1,kAm1],[0,100], style = :dash, color = :black)
plot!([kAlq1,kAlq1],[0,100], style = :dash, color = :black)
plot!([kAuq1,kAuq1],[0,100], style = :dash, color = :black)
```

```julia
a2 = stephist(kB1, fill = true, fillalpha = 0.5, color = cpal[3], label = :none, norm = true, titlelocation = :left)
plot!(grid = false, minorgrid = false, legend = :none, size = (450,400), ylims = [0,0.18],xlims = [2,20])
plot!([kBm1,kBm1],[0,100], style = :dash, color = :black)
plot!([kBlq1,kBlq1],[0,100], style = :dash, color = :black)
plot!([kBuq1,kBuq1],[0,100], style = :dash, color = :black)
```

```julia
a3 = stephist(kC1, fill = true, fillalpha = 0.5, color = cpal[7], label = :none, norm = true, titlelocation = :left)
plot!(grid = false, minorgrid = false, legend = :none, size = (450,400), ylims = [0,0.006],xlims = [0,1000])
plot!([kCm1,kCm1],[0,100], style = :dash, color = :black)
plot!([kClq1,kClq1],[0,100], style = :dash, color = :black)
plot!([kCuq1,kCuq1],[0,100], style = :dash, color = :black)
```

```julia
row1 = plot(a1,a2,a3,layout = (1,3),size = (800,200), bottom_margin = 2mm, top_margin = 0mm)
```
